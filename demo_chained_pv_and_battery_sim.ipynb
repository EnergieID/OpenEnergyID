# Chained PV and Battery Simulation Demo

This notebook is a demo of a "chained simululation", where the output of one simulation (PV) is used as input for another simulation (Battery).

This is achieved by having `SimulationParameters` be a list, and running each simulation sequentially, passing the output of one as input to the next. The results of each simulation are returned in a dictionary, with the keys being the type of the simulation.

```python
from openenergyid.sim import FullSimulationInput, run_simulation
```

```python
with open("data/sim/chained_input.json", "r") as f:
    input_ = FullSimulationInput.model_validate_json(f.read())
```

```python
input_.simulation_parameters
```

```python
result = await run_simulation(input_, session=None)
```

```python
print(result.model_dump_json(indent=2))
```
