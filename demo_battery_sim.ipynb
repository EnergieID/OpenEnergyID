# Battery Simulation

```python
from openenergyid.sim import FullSimulationInput
```

```python
with open("data/sim/batterysim_selfconsumption.json", "r") as f:
    input = FullSimulationInput.model_validate_json(f.read())
```

```python
from openenergyid.battsim.selfconsumption.main import SelfConsumptionBatterySimulator
```

```python
simulator = SelfConsumptionBatterySimulator(
    data=input.ex_ante_data.to_pandas(timezone=input.timezone),
    capacity=5,
    power=5,
    result_resolution="MS",
)
```

```python
simulator.result_to_timedataframe().to_pandas(timezone="Europe/Brussels").round(3)
```
