```python
from openenergyid.sim import FullSimulationInput
```

```python
with open("data/sim/pvlib_quickscan.json", "r") as f:
    input = FullSimulationInput.model_validate_json(f.read())
```

```python
input_data = input.ex_ante_data.to_pandas(timezone="Europe/Brussels")[
    ["electricity_delivered", "electricity_exported"]
]
```

```python
input_data
```

```python
from openenergyid.battsim.selfconsumption.main import SelfConsumptionBatterySimulator
```

```python
simulator = SelfConsumptionBatterySimulator(
    data=input_data, capacity=5, power=5, result_resolution="YS"
)
```

```python
simulator.simulate()
```

```python
input_data.resample("YS").sum()
```

```python
simulator.result_to_timedataframe().to_pandas(timezone="Europe/Brussels").round(3)
```
